```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
missing_values = "NaN"
titanic = pd.read_csv("/content/titanic.csv")
```

```python
print("First 5 rows:")
print(titanic.head())
print("\nMissing values:")
print(titanic.isnull().sum())
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)
titanic.fillna(method='ffill', inplace=True)
```

```python
survival_rate = titanic.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()
```

```python
print("\nSurvival Rate by Pclass and Sex:")
print(survival_rate)
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=survival_rate)
```

```python
plt.title("Survival Rate by Passenger Class and Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.hist(titanic['Age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(y=titanic['Fare'])
plt.title("Box Plot of Fare")
plt.ylabel("Fare")
plt.show()
```
